```python
import cv2
import numpy as np
img=cv2.imread("new_hi.jpg")   #reading image
img.shape # return shape of image x aaxis y axis colors channels
```

```python
img[0][0] # return first row of image matrix
import matplotlib.pyplot as plt
plt.imshow(img) #return or plot image i 2 channels
while True:
    cv2.imshow("Image Window",img)
    if cv2.waitKey(50)==27:  # control to break the loop 
        break

cv2.destroyAllWindows() #destroying all prevoius windos
```

```python
#we are using voila jones algorithm in this code.
# some haar features
haar_data = cv2.CascadeClassifier('face_classifier.xml')  # method use to read XML files
```

```python
haar_data.detectMultiScale(img)  # buitin method in volia jones alogirithms 
                                 # that able to detect faces and retru their
                                 # corresponding coordinates in form of array
```

```python
# cv2.rectangle(img,(x,y),(w,h),(bgr),border_thisckness)
#in this section we are going to put the rectangles around the faces
while True:
    faces=haar_data.detectMultiScale(img) # getting coordinates of detected faces in list
    for x,y,w,h in faces: # spling founded face coordinates in main images and split it one by one
        cv2.rectangle(img,(x,y),(x+w,h+y),(255,0,25,),0) # drawing rectangles around them 
    cv2.imshow("page",img) # after drwwing all the rectansgles showing main image
    if cv2.waitKey()==27:
        break;
cv2.destroyAllWindows()
        
```

```python
capture=cv2.VideoCapture(1)
without_mask=[]
count=0
while True:
    flag,img=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces: # spling founded face coordinates in main images and split it one by one
            cv2.rectangle(img,(x,y),(x+w,h+y),(255,255,25),0) # drawing rectangles around them 
            face_data=img[y:y+h,x:x+w,:]
            cv2.resize(face_data,(50,50))
            count+=1;
            if(len(without_mask)<100):
                without_mask.append(face_data)
                print(count)
#             print(face_data)
        cv2.imshow("page",img) # after drwwing all the rectansgles showing main image
        key=cv2.waitKey(2)
        if key==27:
            break;
capture.release()
cv2.destroyAllWindows()
np.save("with_mask_data_string.npy",without_mask)
```
